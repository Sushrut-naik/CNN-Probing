```python
import matplotlib.pyplot as plt

def plot_line_graphs(data):
    features = ['shape', 'texture', 'color', 'untrained']
    colors = {'shape': 'blue', 'texture': 'orange', 'color': 'green'}

    fig, axs = plt.subplots(1, len(data), figsize=(20, 6))
    
    for i, (target_feature, values) in enumerate(data.items()):
        if(target_feature == 'untrained'):
            axs[i].set_title(f'{target_feature.capitalize()}', fontsize=16, fontweight='bold')
        else:
            axs[i].set_title(f'target feature = {target_feature.capitalize()}', fontsize=16, fontweight='bold')

#         axs[i].set_title(f'Accuracy Comparison for {target_feature.capitalize()}', fontsize=16, fontweight='bold')

        for feature, accuracies in values.items():
            axs[i].plot(list(accuracies.keys()), [acc[1]*100 for acc in accuracies.values()], marker='o', label=feature, linewidth=4, markersize=8, color=colors.get(feature, 'black'))

        axs[i].axhline(y=14.3, color='r', linestyle='--', label='Baseline (14.3%)', linewidth=2)

        axs[i].set_xlabel('Layer', fontsize=14, fontweight='bold')
        axs[i].set_ylabel('Accuracy', fontsize=14, fontweight='bold')
        axs[i].set_yticks(range(0, 101, 20))
        axs[i].tick_params(axis='both', which='major', labelsize=12)
        axs[i].set_ylim(0, 105)
        axs[i].legend(fontsize=12)
        axs[i].grid(True)
    
    plt.tight_layout()
    plt.show()

# Example usage:
# plot_line_graphs(data)
```
